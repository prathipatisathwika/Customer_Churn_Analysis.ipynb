```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
```

```python
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
df["customerID"].duplicated().sum()
```

```python
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
# count of customers who have churned or not churned
import seaborn as sns
ax = sns.countplot( x = "Churn" , data = df )
ax.bar_label( ax.containers[0] )
plt.title("count of customer by churn")
plt.show()
```

```python
# List of columns
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create a 3x3 subplot
fig, axes = plt.subplots(3, 3, figsize=(18, 15))

# Flatten the axes array
axes = axes.flatten()

# Loop through columns and create countplots
for i, col in enumerate(cols):
    sns.countplot(data=df, x=col, ax=axes[i] , palette="viridis" , hue = "Churn" )
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

# Adjust spacing
plt.tight_layout()

# Display the plots
plt.show()
```

**Overall Business Strategy**
Instead of treating all customers the same, the company should identify high-risk customers and target them with personalized retention campaigns.

Examples include:
Customers with tenure less than 12 months should receive onboarding emails, welcome offers, and regular follow-ups.
Customers on month-to-month contracts should be offered incentives to upgrade to annual plans before they consider switching.

**Additional Recommendations**
Build a churn prediction model using machine learning to identify customers likely to leave.
Send personalized offers to customers predicted to churn.
Conduct exit surveys to understand the real reasons behind cancellations.
Monitor customer satisfaction (CSAT/NPS) regularly.
Improve service quality in regions where churn is unusually high.
Create loyalty rewards for long-term customers.
Bundle complementary services (Internet + Phone + Security + Tech Support) at discounted prices to increase customer stickiness.

**One important caution**
Some of these observations show correlation, not necessarily causation. For example, "customers without tech support churn more" does not automatically mean that adding tech support will reduce churn. It could be that customers who are already more engaged are both more likely to buy tech support and more likely to stay.
